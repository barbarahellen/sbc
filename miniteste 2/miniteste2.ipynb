```python
import nltk
from nltk import word_tokenize
nltk.download('punkt')
```

```python
def preprocess(d):
    d = d.lower()
    d = "eos " + d
    d = d.replace("."," eos")
    return d
```

```python
# divide o texto em tokens
def generate_tokens(d):
    tokens = word_tokenize(d)
    return tokens
```

```python
# gera a frequência dos tokens
def generate_tokens_freq(tokens):
    dct = {}    
    
    for i in tokens:
        dct[i] = 0
    for i in tokens:
        dct[i] += 1
    return dct
```

```python
# gera os bigramas (pares de palavras)
def generate_ngrams(tokens,k):
    l = []
    i = 0
    
    while(i<len(tokens)):
        l.append(tokens[i:i+k])
        i = i+1
    l =l[:-1]
    return l
```

```python
# gera a frequência dos bigramas
def generate_ngram_freq(ngrams):
    dct1 = {}
    
    for i in bigram:
        st =" ".join(i)
        dct1[st] = 0
    for i in bigram:
        st =" ".join(i)
        dct1[st] += 1
    return dct1
```

```python
# busca a frequência de um bigrama específico
def find1(s,dct1):
    try:
        return dct1[s]
    except:
        return 0

# calcula a probabilidade de cada bigrama
def print_probability_table(distinct_tokens,dct,dct1,V):
    # para cada par de palavras (word1, word2), calcula-se a probabilidade de P(word2 | word1),
    # usando a frequência do bigrama e da word1 no corpus, ajustadas pelo smoothing.

    n = len(distinct_tokens)
    l = [[0] * n for _ in range(n)]           # n = num de tokens distintos

    for i in range(n):           # i representa a word1
        for j in range(n):       # j representa a word2                                  
            bigram = distinct_tokens[i] + " " + distinct_tokens[j]
            numerator = find1(bigram, dct1) + 1                       # adicionando 1 para o smoothing
            denominator = dct.get(distinct_tokens[i], 0) + V          # adicionando V para o smoothing        
            l[i][j] = float("{:.3f}".format(numerator / denominator))

    return l
```

```python
# número de tokens distintos
distinct_tokens = list(set(sorted(tokens)))
V = len(distinct_tokens)

# tabela de probabilidades das frequências dos tokens
print("Probability table = \n")
probability_table=print_probability_table(distinct_tokens,dct,dct1,V)

n = len(distinct_tokens)
print("\t", end="")
for i in range(n):
    print(distinct_tokens[i],end="\t")
print("\n")
for i in range(n):
    print(distinct_tokens[i],end="\t")
    for j in range(n):
        print(probability_table[i][j],end="\t")
    print("\n")

print("\n", "-"*100)
```
