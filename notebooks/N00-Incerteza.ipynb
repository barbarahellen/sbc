```python
%pip install numpy matplotlib scipy


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Prior belief (uniform distribution)
prior = np.ones(1001) / 1001  # Bias values from 0 to 1 (step size = 0.001)
bias_values = np.linspace(0, 1, 1001)

# Observed data
num_heads = 7
num_flips = 10

# Likelihood function (binomial)
likelihood = binom.pmf(num_heads, num_flips, bias_values)

# Calculate the unnormalized posterior
unnormalized_posterior = likelihood * prior

# Normalize the posterior
posterior = unnormalized_posterior / np.sum(unnormalized_posterior)

# Plotting the posterior distribution
plt.figure(figsize=(8, 5))
plt.plot(bias_values, posterior, label='Posterior Distribution')
plt.xlabel('Bias')
plt.ylabel('Probability Density')
plt.title('Bayesian Posterior Distribution for Coin Bias')
plt.legend()
plt.show()
```
